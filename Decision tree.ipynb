```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier,plot_tree

df=pd.read_csv('https://gist.githubusercontent.com/DiogoRibeiro7/c6590d0cf119e87c39e31c21a9c0f3a8/raw/4a8e3da267a0c1f0d650901d8295a5153bde8b21/PlayTennis.csv')
print(df)
le=LabelEncoder()
df['outlook_n']=le.fit_transform(df['Outlook'])
df['temp_n']=le.fit_transform(df['Temperature'])
df['humidity_n']=le.fit_transform(df['Humidity'])
df['windy_n']=le.fit_transform(df['Wind'])
df['play_n']=le.fit_transform(df['Play Tennis'])

print(df)

df=df.drop(['Outlook','Temperature','Humidity','Wind','Play Tennis'],axis='columns')
print(df)

iv=df.drop('play_n',axis=1)
dv=df['play_n']
model=tree.DecisionTreeClassifier()
X_train, X_test, y_train, y_test = train_test_split(iv,dv,test_size=0.2)

model.fit(X_train,y_train)
model.score(X_train,y_train)
model.predict(X_test)
model.predict([[1,2,0,1]])
plot_tree(model,filled=True,feature_names=df.columns,class_names=le.classes_)
plt.show()
```
