```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Your data
x = np.array([3, 4, 5, 7, 8, 5, 8, 9, 3, 3, 2, 5])
y = np.array([7, 6, 5, 5, 4, 1, 3, 3, 7, 5, 4, 9])
data = np.column_stack((x, y))

dbscan = DBSCAN(eps=1.9, min_samples=4)

dbscan.fit(data)

cluster_labels = dbscan.labels_

for label in np.unique(cluster_labels):
    if label == -1:
        # Noise points
        plt.scatter(data[cluster_labels == label, 0], data[cluster_labels == label, 1], label=f'Noise', color='gray', s=100)
    else:
        # Core and border points
        plt.scatter(data[cluster_labels == label, 0], data[cluster_labels == label, 1], label=f'Cluster {label}', s=100)

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
